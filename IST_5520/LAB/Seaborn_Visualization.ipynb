```python
# Import libraries
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Distribution Analysis

```python
# Histogram
sns.set_style("whitegrid")
tips = sns.load_dataset("tips")
ax = sns.histplot(tips["total_bill"])
```

```python
# Density plot
ax = sns.kdeplot(tips["total_bill"])
```

```python
# Histogram + kernel density estimate
ax = sns.displot(tips["total_bill"], kde = True)
```

```python
# Boxplot
ax = sns.boxplot(x=tips["total_bill"])
```

```python
ax = sns.boxplot(x="day", y="total_bill",hue="time", data=tips);
```

```python
# Violinplot
ax = sns.violinplot(x=tips["total_bill"])
```

# 2. Comparative Analysis

```python
# Show standard deviation of observations
ax = sns.barplot(x="tip", y="day", ci='sd',data=tips)
```

```python
# Add "caps" to the error bars
ax = sns.barplot(x="day", y="tip", ci='sd', data=tips, capsize=.2)
```

```python
g = sns.catplot(x="sex", y="total_bill",
                   hue="smoker", col="time",
                   data=tips, kind="bar",
                   height=4, aspect=.7);
```

# 3. Correlation Analysis

```python
# Load the tips dataset
tips = sns.load_dataset("tips")
```

```python
# Info of the dataset
tips.info()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
# Hexbin plot

mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])

x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

```python
# Scatterplot matrix
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species")
```
