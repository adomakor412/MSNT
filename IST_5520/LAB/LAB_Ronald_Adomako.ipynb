### c: Barchart of Fuel type prices.

```python
#plt.bar?
plt.bar(df['Fuel_Type'],df['Price']);
#matplotlib inline
plt.title('Fuel type prices');
plt.xlabel('Type');
plt.xlabel('Price');
```

### d: Box plot for each fuel type

```python
#df['Fuel_Type'].unique()
```

```python
for fuel in df['Fuel_Type'].unique():
    
    #plt.box?
    #print(fuel)
    'sub data frame'
    #print(df[df['Fuel_Type']==fuel])
    'True/ False'
    #print(df['Fuel_Type']==fuel)
    
    #print(np.array(df[df['Fuel_Type']==fuel]['Price']))
    plt.boxplot( np.array(df[df['Fuel_Type']==fuel]['Price']) )
    plt.title(f'Fuel type: {fuel}');
    #plt.xlabel(f'Type')
    plt.xlabel(f'Price')
    
    #plt.close()
    'Allow all to display when using loop'
    plt.show()
plt.close()
```

### e: Average price of car per fuel type

```python
for fuel in df['Fuel_Type'].unique():
    prices = df[df['Fuel_Type']==fuel]['Price']
    #print(prices)
    #print(np.array(prices))
    mean = np.array(prices).mean()
    print(f'Type: {fuel}, mean: ${np.round(mean,2)}')
```

### a: Price vs Mileage of used corolla

```python
plt.scatter(df['Price'],df['KM']);
plt.title('Price vs. Mileage')
plt.ylabel('Mileage (Km)')
plt.xlabel('Price ($)');
```

### b: Weak negative trend: Price decrease linearly with increasing mileage. 

#### There is a higher correlation of a negative trend when comparing price with mileage per fuel type for Petrol.

```python
for fuel in df['Fuel_Type'].unique():
    fuel_df = df[df['Fuel_Type']==fuel]
    plt.scatter(fuel_df['Price'],fuel_df['KM'])
    plt.title(f'{fuel} Price vs. Mileage')
    plt.ylabel(f'Mileage (Km)')
    plt.xlabel(f'Price ($)')
    plt.show()
```
