## (7) Draw a bar chart of CCAvg for all family sizes.

Show a bar for each family size.

Show standard errors on the bars.

```python
df_fam = df[['Family','CCAvg']]
df_fam
```

```python
meanFam = df[['Family','CCAvg']].groupby('Family').mean()
#meanFam.columns = ['CCAvg_mean']
meanFam
```

```python
std1 = df[['Family','CCAvg']].groupby('Family').std()
#std1.columns = ['CCAvg_std']
#std1.index = ['1','2','3','4']
std1
```

```python
meanFam.plot(kind='bar', title='Universal Bank: Family and errs',
             align = 'center',
             yerr = std1,
             legend = False,
             xlabel='Family Size',
             ylabel = 'Average CCAvg');
plt.xticks(rotation=45);

```
