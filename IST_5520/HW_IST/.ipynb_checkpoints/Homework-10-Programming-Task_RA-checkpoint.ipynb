```python
# df_category_OHE = pd.get_dummies(df_category, prefix = 'OHE')
# df_category_OHE
df_var_OHE = df_var.drop(['season','holiday','workingday','weather'],axis=1)
df_var_OHE
```

```python
for ohe in df_category.columns:
    df_ohe = pd.get_dummies(df_var[ohe], prefix=f'{ohe}')
    #print(df_ohe.columns)
    #frames = [df_ohe[[c]] for c in df_ohe.columns]
    #print(frames[0])
    #frames.insert(0, df_var_OHE)
    #print(frames)
    for c_ohe in df_ohe.columns:
        df_var_OHE[c_ohe] = df_ohe[c_ohe].astype(float)
df_var_OHE
```

3. Use 40-60% data partition strategy, with 40% of data used as test data. 

```python
X_train, X_test, y_train, y_test = train_test_split(
  df_var_OHE, df_target, test_size=0.4, random_state=42)
```

2. Use k-NN algorithm to predict the count of bike rental. Tune the parameter of k based on the RMSE statistic.


Note: As this is a regression problem, you’ll need to use an appropriate method. Refer to the
link for all methods offered in the skearn.neighbors module:


https://scikit-learn.org/stable/modules/classes.html#module-sklearn.neighbors

```python
neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(X_scaled, y_train)
```

```python
y_pred = neigh.predict(X_test)
y_pred
```

Compare the performance of the k-NN regressors with different values of parameter k.
Answer the following questions.

Q1: What is the value of k that leads to the best performance of k-NN in terms of RMSE?
Answer:

#### sample RMSE of k=2

```python
mse = sklearn.metrics.mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)
rmse
```

```python
root_mse = []
mae = []
r2 = []
for i in range(2,1000):
    neigh = KNeighborsRegressor(n_neighbors=i);
    neigh.fit(X_scaled, y_train);
    y_pred = neigh.predict(X_test);
    mse = metrics.mean_squared_error(y_test, y_pred);
    rmse = math.sqrt(mse);
    root_mse.append(rmse);
    mae.append(metrics.mean_absolute_error(y_test, y_pred))
    r2.append(metrics.r2_score(y_test, y_pred))
root_mse;
```

```python
plt.plot(range(2,1000),root_mse)
plt.title('KNN RMSE for City bike in DC')
```

```python
plt.plot(range(2,1000),mae)
plt.title('KNN MAE for City bike in DC')
```

```python
plt.plot(range(2,1000),r2)
plt.title('KNN R-square for City bike in DC')
```

```python
print(f'rmse:{min(root_mse)}, k:{root_mse.index(min(root_mse))+2}')

print(f'mae:{min(mae)}, k:{mae.index(min(mae))+2}')

print(f'R-square:{max(r2)}, k:{r2.index(max(r2))+2}')
```
